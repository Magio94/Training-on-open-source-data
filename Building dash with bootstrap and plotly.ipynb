```python
import geopandas as gpd, pandas as pd, matplotlib.pyplot as plt, plotly, numpy as np

import plotly.express as px 
import plotly.graph_objects as go
import pandas_datareader.data as web
import datetime
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objects as go
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output
```

```python
points_within = gpd.read_file('/Users/magio94/Dropbox/Bottom-up/_Tests_area/Tests_ZukunftsmesseWL/Shapefiles/points_within.shp')
municipalities_westfalen_lippe = gpd.read_file('/Users/magio94/Dropbox/Bottom-up/_Tests_area/Tests_ZukunftsmesseWL/Shapefiles/municipalities_westfalen_lippe.shp')
#activity_count = gpd.read_file('/Users/magio94/Dropbox/Bottom-up/_Tests_area/Tests_ZukunftsmesseWL/Shapefiles/activity_count.shp')
```

```python
dff = points_within.copy()
dff2 = municipalities_westfalen_lippe.copy()

# https://www.bootstrapcdn.com/bootswatch/
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.CYBORG],
                meta_tags=[{'name': 'viewport',
                            'content': 'width=device-width, initial-scale=1.0'}]
                )

app.layout = dbc.Container([

    dbc.Row(
        dbc.Col(html.H1("Different activities by category",
                        className='text-center bg-gradient-primary text-white mb-4'),
                width=12)
    ),
    
    dbc.Row([
        dbc.Col([
            html.H4("Map divided by categorical activities",
                    className='text-center bg-gradient-primary text-white mb-4'),
            
            dcc.Dropdown(id="category_selector",
                         options=[
                             {'label': i, 'value': i} for i in dff['category'].unique()],
                         value=['Association', 'NPO'],
                         multi=True,
                             ),
            dcc.Graph(id='map')],
            width={'size':5, 
                   'offset':1
                  }),       
    
        dbc.Col([dcc.Dropdown(id="dropdown",
                              options=[
                                  {'label': n, 'value': n} for n in dff['NAME_2'].unique()],
                              value = 'Arnsberg',
                              multi=False,
                             ),
                
                dcc.Graph(id='bar-chart')],
                className= 'text-center bg-secondary text-white mb-4',
                width={'size':5, 
                       #'offset':1
                      }),      
    ]),
    
], 

    fluid=True)

@app.callback(Output('map', 'figure'),
              [Input('category_selector', 'value')])

def update_graph(option_slctd):
    dff = points_within.copy()
    
    dff = dff[dff['category'].isin(option_slctd)]
    dff2 = municipalities_westfalen_lippe.copy()

    fig = px.choropleth_mapbox(dff2,
                               geojson=dff2.geometry,
                               locations=dff2.index,
                               color="WL_provinc",
                               mapbox_style="open-street-map",
                               opacity=0.5,
                               zoom=6,
                               center = {"lat": 51.71027, "lon": 8.059108}
                              )
    fig.update(layout_showlegend=False)

    fig2 = px.scatter_mapbox(dff,
                             lat="Lat",
                             lon="Lon",
                             # hover_name= "name",
                             # color_discrete_sequence=["fuchsia"],
                             color=dff.category,
                             # hover_data = ["name"]
                             )

    fig2.update_layout(mapbox_style="white-bg")
    fig.update(layout_showlegend=True)

    for x in range(0, len(option_slctd)):
        fig.add_trace(fig2.data[x])


    return fig

@app.callback(
    Output("bar-chart", "figure"), 
    [Input("dropdown", "value")])

def update_bar_chart(NAME_2):
    
    dff2 = dff[dff['NAME_2'] == NAME_2]
    
    fig = px.bar(dff2.category, y= dff2.NAME_3)
    fig.update_layout(yaxis={'categoryorder':'total descending'})
    
    return fig

if __name__=='__main__':
    app.run_server(debug=False, port=8000)
```
